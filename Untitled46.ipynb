```python
#write a python class representing a discrite random variable with method to calculate its expected value and variance

class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize a discrete random variable with outcomes and corresponding probabilities.

        outcomes: list or tuple of possible outcomes
        probabilities: list or tuple of corresponding probabilities
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("Number of outcomes must be equal to number of probabilities.")
        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if abs(sum(probabilities) - 1.0) > 1e-10:
            raise ValueError("Probabilities must sum to 1.")

        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the random variable.
        """
        expected_val = sum(x * p for x, p in zip(self.outcomes, self.probabilities))
        return expected_val

    def variance(self):
        """
        Calculate the variance of the random variable.
        """
        mean = self.expected_value()
        variance = sum(p * (x - mean)**2 for x, p in zip(self.outcomes, self.probabilities))
        return variance

# Example usage:
outcomes = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

# Create an instance of DiscreteRandomVariable
rv = DiscreteRandomVariable(outcomes, probabilities)

# Calculate and print expected value and variance
print("Expected value:", rv.expected_value())
print("Variance:", rv.variance())
```

```python
# implemented a program to simulated the rolling of a fair six-added die and calculated the expected value and varience of the outcomes.
import random

class FairSixSidedDie:
    def __init__(self):
        self.sides = 6

    def roll(self):
        """
        Simulate rolling the die and return the result (a number between 1 and 6).
        """
        return random.randint(1, self.sides)

    def expected_value(self):
        """
        Calculate the expected value (mean) of rolling the die.
        For a fair die, E[X] = (1 + 2 + 3 + 4 + 5 + 6) / 6 = 3.5
        """
        return (1 + 2 + 3 + 4 + 5 + 6) / 6

    def variance(self):
        """
        Calculate the variance of rolling the die.
        For a fair die, Var(X) = E[X^2] - (E[X])^2
        """
        expected_val = self.expected_value()
        expected_val_squared = sum(x**2 for x in range(1, self.sides + 1)) / self.sides
        return expected_val_squared - expected_val**2

# Create an instance of FairSixSidedDie
die = FairSixSidedDie()

# Simulate rolling the die multiple times to estimate expected value and variance
num_simulations = 100000
results = [die.roll() for _ in range(num_simulations)]

# Calculate empirical expected value
empirical_expected_value = sum(results) / num_simulations

# Calculate empirical variance
empirical_variance = sum((result - empirical_expected_value)**2 for result in results) / num_simulations

# Compare with theoretical values
theoretical_expected_value = die.expected_value()
theoretical_variance = die.variance()

print(f"Theoretical Expected Value: {theoretical_expected_value}")
print(f"Theoretical Variance: {theoretical_variance}")
print(f"Empirical Expected Value (from {num_simulations} simulations): {empirical_expected_value}")
print(f"Empirical Variance (from {num_simulations} simulations): {empirical_variance}")
```

```python
#5.create random a python function to generated random sample from a given probality distribution and calcuated their mean and variance
import numpy as np

def generate_random_samples(distribution, size=1000):
    """
    Generate random samples from a given probability distribution and calculate their mean and variance.

    Parameters:
    - distribution: NumPy probability distribution object (e.g., np.random.normal, np.random.binomial, etc.)
    - size: Number of samples to generate (default is 1000)

    Returns:
    - mean: Mean of the generated samples
    - variance: Variance of the generated samples
    """
    # Generate random samples
    samples = distribution(size=size)

    # Calculate mean and variance
    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance

# Example usage:
mean, variance = generate_random_samples(np.random.normal)  # Generates samples from a normal distribution
print(f"Mean of samples: {mean}")
print(f"Variance of samples: {variance}")
```

```python
#8.write a python function to calculated the probability density function (PDF) of a continuouse random variable for a given normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def normal_pdf(x, mu, sigma):
    """
    Compute the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    x (float or numpy array): The point(s) at which to compute the PDF.
    mu (float): Mean of the normal distribution.
    sigma (float): Standard deviation of the normal distribution.

    Returns:
    float or numpy array: PDF value(s) corresponding to x.
    """
    return norm.pdf(x, loc=mu, scale=sigma)

# Example usage:
mu = 0    # Mean of the normal distribution
sigma = 1 # Standard deviation of the normal distribution

# Generate x values for plotting
x = np.linspace(-5, 5, 100)  # Range of x values

# Compute the PDF values for the given normal distribution
pdf_values = normal_pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, pdf_values, label=f'Normal PDF (μ={mu}, σ={sigma})')
plt.title('Normal Distribution Probability Density Function')
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9.create a program to calculated the cumulatiove distribution function (CDF)to exponetial distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

def exponential_cdf(x, scale):
    """
    Compute the Cumulative Distribution Function (CDF) of an exponential distribution.

    Parameters:
    x (float or numpy array): The point(s) at which to compute the CDF.
    scale (float): Scale parameter (inverse of the rate parameter) of the exponential distribution.

    Returns:
    float or numpy array: CDF value(s) corresponding to x.
    """
    return expon.cdf(x, scale=scale)

# Example usage:
scale_parameter = 1.0  # Scale parameter of the exponential distribution

# Generate x values for plotting
x = np.linspace(0, 5, 100)  # Range of x values

# Compute the CDF values for the exponential distribution
cdf_values = exponential_cdf(x, scale=scale_parameter)

# Plot the CDF
plt.plot(x, cdf_values, label=f'Exponential CDF (λ={1/scale_parameter})')
plt.title('Exponential Distribution Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#10.write a python function to calculate the probability mass function (PMF) of possion distribution
import math

def poisson_pmf(k, lambd):
    """
    Compute the Probability Mass Function (PMF) of a Poisson distribution.

    Parameters:
    k (int): Number of occurrences.
    lambd (float): Mean parameter (λ) of the Poisson distribution.

    Returns:
    float: PMF value for P(X = k).
    """
    return (lambd ** k) * math.exp(-lambd) / math.factorial(k)

# Example usage:
lambd = 2.5  # Mean parameter of the Poisson distribution

# Calculate and print PMF for various values of k
for k in range(0, 10):
    pmf_value = poisson_pmf(k, lambd)
    print(f"PMF at k={k}: {pmf_value}")
```

```python
#11.A company wants to test if a new website layout leads to a higher conversion rate (percentege of visitor who make a purchase ) .They collect data from the old and new layout to compare

to generate the dataset data use the following command:
'''python
import numpy as np
#50 purchase out of 1000 visitor
old_layout= np.array([1] * 50 + [0] * 950)
#70 purchase out of 1000 visitor
new_layout = np.array[1] * 70 + [0] * 930)
'''

#apply z test to find which layout is successful

import numpy as np
from scipy import stats

# Generate data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Calculate proportions and sample sizes
n_old = len(old_layout)
n_new = len(new_layout)
p_old = np.mean(old_layout)
p_new = np.mean(new_layout)

# Calculate standard error of the difference in proportions
SE = np.sqrt(p_old * (1 - p_old) / n_old + p_new * (1 - p_new) / n_new)

# Calculate Z-score
Z = (p_new - p_old) / SE

# Two-tailed test, calculate p-value
p_value = stats.norm.cdf(Z)

# Print results
print(f"Z-score: {Z:.4f}")
print(f"P-value: {p_value:.4f}")

# Check significance at alpha = 0.05
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)  # One-sided test for upper tail

print(f"Critical value (alpha={alpha}): {critical_value:.4f}")

if Z > critical_value:
    print("Reject null hypothesis: The new layout leads to a higher conversion rate.")
else:
    print("Fail to reject null hypothesis: There is no sufficient evidence that the new layout leads to a higher conversion rate.")

```

```python
#15.A company is testing two different website layout to see which one leads to higher click-through rates.write a python function to perform an A/B analysis, including calculating the t-statistics, degrees of freedom and p-value.

use the following data
''' python
layout_a_clicks = [28,32,33,29,31,34,30,35,36,37]
layout_b_clcks = [40,41,38,42,39,44,43,41,45,47]

'''

import numpy as np
from scipy import stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Calculate sample statistics for layout A
    mean_a = np.mean(layout_a_clicks)
    std_dev_a = np.std(layout_a_clicks, ddof=1)  # using ddof=1 for sample std deviation
    n_a = len(layout_a_clicks)

    # Calculate sample statistics for layout B
    mean_b = np.mean(layout_b_clicks)
    std_dev_b = np.std(layout_b_clicks, ddof=1)
    n_b = len(layout_b_clicks)

    # Calculate t-statistic
    pooled_std_dev = np.sqrt((std_dev_a**2 / n_a) + (std_dev_b**2 / n_b))
    t_stat = (mean_a - mean_b) / pooled_std_dev

    # Calculate degrees of freedom
    df = n_a + n_b - 2

    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))  # two-tailed test

    return t_stat, df, p_value

# Data provided
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
t_stat, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

print(f"T-statistic: {t_stat}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")
```

```python
#17.A school discrict introduce an educational intervention program to improve math score.write a python function to analyze pre-and post-intervenion test scores,calculating the t-statistics and p-value to determine if the intervention had a significant impact.

use the following data of test score:

''' python
pre_intervention_scores = [80,85,90,75,88,82,92,78,85,87]
post_intervention_scores = [90,92,88,92,95,91,96,93,89,93]
'''

import numpy as np
from scipy import stats

def intervention_effect(pre_intervention_scores, post_intervention_scores):
    # Calculate differences
    differences = np.array(post_intervention_scores) - np.array(pre_intervention_scores)

    # Calculate sample statistics
    mean_diff = np.mean(differences)
    std_dev_diff = np.std(differences, ddof=1)
    n = len(differences)

    # Calculate t-statistic
    t_stat = mean_diff / (std_dev_diff / np.sqrt(n))

    # Calculate degrees of freedom
    df = n - 1

    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))  # two-tailed test

    return t_stat, df, p_value

# Data provided
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform analysis
t_stat, df, p_value = intervention_effect(pre_intervention_scores, post_intervention_scores)

print(f"T-statistic: {t_stat}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")
```

```python
#19.a manufacture produces two different version of product and wants to compare their quality scores.craete a python function to analyze quality asseement data,calculate the t-statistics , and decide whether there a significant difference in quality between the two versions.

use the following data:

''' python
versiona1_scores = [85,88,82,89,87,84,90,88,85,91,83,87,84,86,84,88,85,89,90,87,88,85]
version2_scorea  = [80,78,83,81,79,82,76,80,78,81,77,82,80,79,82,79,80,81,79,82,79,78,80,81,82]
'''
import numpy as np
from scipy import stats

def compare_quality(version1_scores, version2_scores):
    # Calculate sample statistics for version 1
    mean_v1 = np.mean(version1_scores)
    std_dev_v1 = np.std(version1_scores, ddof=1)
    n_v1 = len(version1_scores)

    # Calculate sample statistics for version 2
    mean_v2 = np.mean(version2_scores)
    std_dev_v2 = np.std(version2_scores, ddof=1)
    n_v2 = len(version2_scores)

    # Calculate t-statistic
    pooled_std_dev = np.sqrt((std_dev_v1**2 / n_v1) + (std_dev_v2**2 / n_v2))
    t_stat = (mean_v1 - mean_v2) / pooled_std_dev

    # Calculate degrees of freedom
    df = n_v1 + n_v2 - 2

    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))  # two-tailed test

    return t_stat, df, p_value

# Data provided
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 91, 83, 87, 84, 86, 84, 88, 85, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

# Perform comparison
t_stat, df, p_value = compare_quality(version1_scores, version2_scores)

print(f"T-statistic: {t_stat}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")
```

```python
#21.A political analyst wants to determine if there is a significant association between age group and voter preference(Candiate A or Candidate B).They collect data from a sample 500 voters and classify them into different age group and candiate prefernce.perform a chi-Square test to determine if there is a significant association between age group and voter prefrences.

use the below code to generate data :
'''python
np.random.seed(0)
age_group = np.random.choice(['18-30','31-50' ,'51+','51+'],size = 30)
voter_preference = np.random.choice(['Cadidate A','Candidate B'],size = 30)
'''
import numpy as np
from scipy.stats import chi2_contingency

np.random.seed(0)
age_group = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preference = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create contingency table
contingency_table = np.zeros((3, 2), dtype=int)

# Populate contingency table
for i in range(len(age_group)):
    if age_group[i] == '18-30':
        row = 0
    elif age_group[i] == '31-50':
        row = 1
    else:
        row = 2

    if voter_preference[i] == 'Candidate A':
        col = 0
    else:
        col = 1

    contingency_table[row, col] += 1

print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test of independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("\nExpected Frequencies:")
print(expected)
```
